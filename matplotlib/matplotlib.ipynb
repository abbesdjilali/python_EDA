### Une certaine m√©thodologie  

Comme beaucoup de librairies il y a une certaine logique √† saisir üßê 

`matplotlib` vous offre deux logiques de construction de graphiques. Une fa√ßon rapide via la m√©thode `plot` et une **fa√ßon objet** via `subplots` 


#### Voyons pour l'instant la m√©thode `plot()` de `matplotlib.pyplot`. 

On utilisera souvent le terme "ploter" ou "plot" en r√©f√©rence √† cette m√©thode qui veut dire tracer un graphique. 

Cette m√©thode nous permet de construire un graphique et de lui ajouter des choses par la suite comme des noms d'axes, des couleurs et un titre par exemple üöÄ    


````python
plt.plot(x, x)
plt.xlabel('axe des x')
plt.ylabel('axe des y');
````

On ajoute un titre √† notre graphique avec la fonction `plt.title()` üî§

````python
plt.plot(x, x, c='green')
plt.xlabel('axe des x')
plt.ylabel('axe des y')
plt.title('Un titre de graphique');
````

On peut aussi ajouter une autre courbe sur le m√™me graphique, simplement en la pla√ßant √† la suite üì∂

````python
plt.plot(x, x)
plt.plot(x, x**3)
plt.xlabel('axe des x')
plt.ylabel('axe des y')
plt.title('Un graphique avec plusieurs courbes');
````

Deux courbes c'est bien mais c'est une information en plus, c'est mieux de l'expliciter avec une l√©gende ü§ì

````python
plt.plot(x, x**2,c="red", label='y=x^2')
plt.plot(x, x**3,c="green", label='y=x^3')
plt.xlabel('axe des x')
plt.ylabel('axe des y')
plt.title('Un graphique avec encore plusieurs courbes + une l√©gende')
plt.legend();
````

Comme vous pouvez le remarquer la m√©thode `legend()` fait r√©f√©rence √† l'argument `label` dans les fonctions `plot`, si vous ne le mettez pas de l√©gende üôÑ 

On peut vouloir afficher uniquement une partie des axes `x` et `y`, on utilise pour cela les fonctions `xlim()` et `ylim()`. 


````python
plt.plot(x, x)
plt.xlim([0.0,1.2])
plt.ylim([0.0,2])
plt.xlabel('axe des x plus petit')
plt.ylabel('axe des y plus petit');
````

#### La m√©thode `subplot()` 

La m√©thode `subplot` permet de tracer plusieurs graphiques dans une m√™me `figure()`. Il faut juste pr√©c√©der vos `plot` de `subplot(lignes,colonnes,position)`. 

````python
plt.figure(figsize=(8,4))
plt.subplot(1,3,2)
plt.title('une fonction lin√©aire')
plt.plot(x,x,c='black')
plt.subplot(1,3,1)
plt.plot(x,x**2, c='red')
plt.title('une fonction quadratique');
plt.subplot(1,3,3)
plt.plot(x,x**3,c="pink")
plt.title("Pink line")
````

On peut remarquer la m√©thode `plt.figure(figsize=(12,6))` qui permet de d'agrandir la figure ü§î

#### Voir vos graphiques comme des objets 

Comme √©nonc√© plus haut, la librairie `matplotlib` nous permet de traiter nos graphiques comme des objets avec la m√©thode `subplots()` qui nous permet de tracer une grille (grids en anglais) de figures. 

On utilisrera la m√©thode `subplots(lignes,colonnes,position)` de la fa√ßon suivante : 

```python 
fig, ax = plt.subplots(2,1,sharex=True)

```

o√π `fig` et `ax` sont respectivement des objets `matplotlib` et `numpy`.  

Le bon c√¥t√© de partager la m√™me figure est qu'on peut partager les m√™mes axes, on utilisera pour cela les arguments `sharex` et `sharey`. 

On va donc voir comment refaire les graphiques pr√©c√©dents avec cette m√©thodes üî≠

````python
fig, ax = plt.subplots(1,1)
ax.plot(x,x)
plt.title(u'une fonction lin√©aire avec subplots');
````

````python
fig, ax = plt.subplots(2,1,sharex=True,figsize=(8,6))
ax[0].plot(x,x)
ax[0].set_title('une fonction lin√©aire')
ax[1].plot(x,x**2)
ax[1].set_title('une fonction quadratique'); 
````

Sachant que `ax` est un `ndarray` vous pouvez voir 
```python
fig, ax = plt.subplots(2,1,sharex=True)
```

√©crit de cette fa√ßon : 
```python
fig, (ax1,ax2) = plt.subplots(2,1,sharex=True)
```

````python
fig, (ax1,ax2) = plt.subplots(1,2,sharex=True,figsize=(8,6))
ax1.plot(x,x)
ax1.set_title(u'une fonction lin√©aire')
ax2.plot(x,x**2)
ax2.set_title('une fonction quadratique'); 
````

Vous pouvez noter que nous n'utilisons plus la m√©thode `plt.title()` mais `set_title()` ü§î

#### Combiner plusieurs graphiques avec  `figure()` et `subplot()`
Dans l'exemple ci-dessous nous utiliserons encore les fonctions polynomiales de base afin "d'encapsuler" plusieurs plots dans une m√™me figure. On d√©taillera les graphiques utilis√©s dans la suite de ce cours. 

$carre(x)=x^2$ et $cubic(x)=x^3$ sont les fonctions que nous utiliserons d√©finies √† l'aide de `fonction lambda` pour la suite de nos exemples ü§ì

````python
carre = lambda x:x**2
cubic = lambda x:x**3
t =[cubic(x) for x in range(10)]
t
````

````python
#Exemple de figure avec des graphes multiples 
plt.figure(1, figsize=(12,8))
plt.subplot(1, 2, 1)
plt.scatter(range(10), [carre(x) for x in range(10)], color = 'blue')
plt.title('figure 1')
plt.subplot(2, 2, 2)
plt.plot(range(-10,10), [cubic(x) for x in range(-10,10)],color = 'red')
plt.title('figure 2')
plt.subplot(2, 2, 4)
plt.bar(range(10), range(10), color = 'green')
plt.title('figure 3'); 
````

L'avantage de cette m√©thode est que nous n'avons pas √† g√©rer le dimentionnement des figures üòé

Combiner des histogrammes permet de visualiser rapidement les distributions de variables üìà

### [Comment tracer des diagrammes circulaire](#pie)

Un Pie chart ou graphique circulaire en fran√ßais souvent appel√© *camembert*, est un type de graphique utilis√© en statistique. Il permet de repr√©senter un petit nombre de valeurs par des angles, o√π la valeur de chaque angle est proportionnelle √† la valeur en pourcentage. 

Voyons dans un premier temps notre dataset `xl` et plus particuli√®rement la variable `Item` üßê

````python
lab = xl.Item.unique(); lab
````

````python
plt.pie(xl.Item.value_counts(), labels=lab);
````

On peut aussi passer par la m√©thode plot de pandas tel que :

````python
# üöß xl.Item.plot(kind='pie') 
xl.Item.value_counts().plot(kind='pie');
````

L'argument `autopct` permet de repr√©senter ces `%`, avec la pr√©cision que vous voulez, les arguments `explode` et `labeldistance` permettent d'ajouter un peu de style √† votre graphique üòé

````python
plt.figure(figsize=(12,8))
plt.pie(xl.Item.value_counts(), labels=xl.Item.unique().tolist(),
        autopct='%.2f',
        explode = [0.2, 0, 0, 0, 0],labeldistance = 1.1);
````

````python
#plus complet 
plt.figure(figsize = (8, 6))
x = [1, 2, 3, 4, 10]
plt.pie(x, labels = ['A', 'B', 'C', 'D', 'E'],
           colors = ['red', 'green', 'yellow'],
           explode = [0, 0.2, 0, 0, 0],
           autopct = lambda x: str(round(x, 2)) + '%',
           pctdistance = 0.7, labeldistance = 1.4,
           shadow = True)
plt.legend();
````

### [Visualiser des nuages de points](#scatter)

Un scatter plot ou nuage de points en fran√ßais est une repr√©sentation de plusieurs variables. Il permet de mettre en √©vidence le *degr√© de corr√©lation* entre deux variables.

Les observations de nuages de points permettent notamment d'analyser :
* Les d√©pendances
* Les relations positives et n√©gatives
* L'homog√©n√©it√© des r√©partitions
* Les donn√©es aberrantes

Dans la suite on utilisera le dataset `list_appart` afin d'analyser le nombre de lits en fonction du nombre de salles de bain par exemple ü§î

````python
plt.figure(figsize=(12,8))
plt.scatter(appart.lits, appart['sale de bain'],c='black')
plt.xlabel('Nombre de lits')
plt.ylabel('Nombre de salles de bain')
plt.title('Nombre de lits en fonction des salles de bain');
````

#### Ajouter des variables √† votre graphique

Il est parfois utile d'ajouter plusieurs variables/dimensions √† vos graphiques afin de mettre en lumi√®re plusieurs groupes/cat√©gories üî≠

````python
plt.figure(figsize=(12,8))
plt.scatter(appart.lits, appart['sale de bain'],c='red',label='sale de bain')
plt.scatter(appart.lits, appart['capacit√©'],c='green',label='capacit√©')
plt.xlabel('Lits')
plt.ylabel('Capacit√© , Sale de bain')
plt.legend()
plt.title('Nombre de lits en fonction de plusieurs param√®tres');
````

On peut avoir envie d'ajouter plus de variables, pour cela  on peut passer par une petite manipulation avec les tuples et les dictionnaires üõ† 

````python
g1 = (appart.lits, appart['sale de bain'])
g2 = (appart.lits, appart['capacit√©'])

data = (g1, g2)
colors = ("red", "green")
groups = ('sale de bain','capacit√©')

plt.figure(figsize=(12,8))
for data, color, group in zip(data, colors, groups):
    x,y = data
    plt.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)
    plt.legend()

plt.xlabel('Lits')
plt.ylabel('Capacit√© , Sale de bain')
plt.title('Nombre de lits en fonction de plusieurs param√®tres');
````

## Tracer des graphiques 3D

Afin de tracer des graphiques 3D, il faut faire appel √† `mplot3d` de `mpl_toolkits`. 

Dans la suite on utilisera deux fonctions $f_1$ et $f_2$, d√©finies √† partir de `fonction lambda` pour illustrer nos exemples ü§ì 

$$f_1(x,y) = sin(x*y)+cos(x)*sin(y)$$
$$f_2(x,y) = e^{(x^2+y^2)}$$

````python
from mpl_toolkits.mplot3d import Axes3D
f1 = lambda x,y : np.sin(x*y)+np.cos(x)*np.sin(y)
f2 = lambda x,y : np.exp(x**2+y**2)
````

````python
z = f1(x,x)
z
````

Afin de passer d'une figure 2D √† une figure 3D on utilise les lignes suivantes 
```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```
Vous noterez l'argument `projection='3d'` assez explicite de la m√©thode `add_subplot()` ü§ì

````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
````

On a bien g√©n√©rer un graphique 3D, maintenant il faut l'accompagner d'une belle courbe ü§ì

On va pour cela cr√©er deux vecteurs `u` et `v` √† l'aide de la m√©thode `linspace`


````python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
````

````python
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111, projection='3d')
plt.plot(u,v,f1(u,v))
plt.title("Courbe sinuso√Ødale en 2D √† partir de tableaux 1D");
````

‚ö†Ô∏è On voit bien les 3 axes et la figue en 3D mais cela ressemble beaucoup √† un plot 2D ‚ö†Ô∏è

En effet, nous avons reli√© les points d'un vecteur 1D en fonction d'un autre vecteur 1D alors que nous avons √† notre disposition deux superbes axes et donc toute une grille de valeurs. Il nous faut donc une grille de point √† tracer ü§î 

Pour cela nous avons deux options qui s'offrent √† nous : 
* Effectuer le [produit tensorielle](https://fr.wikipedia.org/wiki/Produit_dyadique) de nos vecteurs avec la m√©thode `np.outer()`.
* Utiliser la m√©thode `np.meshgrid()` qui g√©n√®re une grille √† partir de vecteurs. 

Nous allons ici utiliser la m√©thode `outer()`, vous pouvez essayer de faire le m√™me travail de votre c√¥t√© avec `meshgrid()` üßê

````python
x = 10 * np.outer(u,v)
y = 10 * np.outer(u,v)
````

````python
f2(u,v).shape
x.shape==y.shape
````

````python
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111, projection='3d')
plt.plot(x,y,f1(u,v))
plt.title("Courbe sinuso√Ødale en 3D √† partir de tableaux 2D");
````
