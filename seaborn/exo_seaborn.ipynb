```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#) Donnez la liste des variables présentes dans ce dataset ainsi que leur
# nature (sont elles ualitatives,
#antitatives, discrète etc…) et leur type (float, int, str etc…) 
house_pricing = pd.read_csv('./data/house_pricing.csv')
house_pricing.info()
```

```python
#En créant un nuage de points, regardez comment se comporte la colonne LotArea par
#rapport au SalesPrices
sns.scatterplot(data = house_pricing, x = "LotArea", y = "SalePrice")
plt.show()
```

```python
#Affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotArea
#inférieur à 20 000 pieds carrés et un prix inférieur à 500 000$ 
result = house_pricing.where((house_pricing["LotArea"] < 20000) & (house_pricing["SalePrice"] < 500000))
sns.scatterplot(data = result, x = "LotArea", y ="SalePrice")
plt.show()
```

```python
#En créant un nuage de points, regardez la relation entre le LotFrontage et le LotArea
sns.scatterplot(data = house_pricing, x = "LotFrontage", y = "LotArea")
plt.show()
```

```python
# De la même manière, affinez votre visualisation en ne gardant uniquement les maisons
# qui ont un LotFrontage inférieur à 200 pieds carrés et un LotArea inférieur à 100000
# pieds carré 
result2 = result.where((result["LotFrontage"] < 200) & (result["LotArea"] < 100000))
sns.scatterplot(data = result2, x = "LotFrontage", y ="LotArea")
plt.show()
```
